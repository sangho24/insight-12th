```python
## 1. initial   (Mr, Miss ) .

df['Initial'] = 0 # initial     0 

for index, row in df.iterrows():
    initial_search = row['Name'].split(',')[1].split('.')[0].strip() # Name  .(dot)    
    df.at[index, 'Initial'] = initial_search

df
```

```python
## 2.    , Mr, Mrs, Miss, Master, Other .

#    ,   Initial Other 

df['Initial'].replace(
    [
        'Mlle', 'Mme', 'Ms', 'Dr', 'Major', 'Lady', 'Countess', 'Jonkheer', 'Col',
        'Rev', 'Capt', 'Sir', 'Don','the Countess' 
    ], 
    [
        'Miss', 'Miss', 'Miss', 'Mr', 'Mr', 'Mrs', 'Mrs', 'Other', 'Other',
        'Other', 'Mr', 'Mr', 'Mr', 'Other'
    ],
    inplace=True)

df
```

```python
# 3.  initial  .

df.loc[(df.Age.isnull()) & (df.Initial == 'Mr'), 'Age'] = 33
df.loc[(df.Age.isnull()) & (df.Initial == 'Mrs'), 'Age'] = 36
df.loc[(df.Age.isnull()) & (df.Initial == 'Master'), 'Age'] = 5
df.loc[(df.Age.isnull()) & (df.Initial == 'Miss'), 'Age'] = 22
df.loc[(df.Age.isnull()) & (df.Initial == 'Other'), 'Age'] = 46
```

##  
     .
- **Sex**  male : 0, female : 1
- **Age**  (: 23 -> 2)
- **Fare** 9 label
- **Embarked** S : 1, C : 2, Q : 3
- **Initial** Mr : 0, Miss : 1, Mrs : 2, Master : 3, Other :4
- **Relatives** : SibSp + Parch

```python
# Sex   male -> 0, female -> 1 . 
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Age  10   .
df['Age'] = (df['Age'] // 10).astype(int)

# Fare  9 .
df['Fare'] = pd.qcut(df['Fare'], q=9, labels=range(9))

# Embarked    .
df['Embarked'] = df['Embarked'].map({'S': 1, 'C': 2, 'Q': 3})

# Initial   .
initial_mapping = {'Mr': 0, 'Miss': 1, 'Mrs': 2, 'Master': 3, 'Other':4}
df['Initial'] = df['Initial'].map(initial_mapping)

df['Relatives'] = df["SibSp"] + df["Parch"]

df
```

#    target feature 
 ,             .   target feature  .
- feature(    =  ):     = Pclass, Sex, Age, SibSp, Parch, Fare, Embarked, Initial, Relatives
- target(   =  ): Survived

 target feature      predict_survival .
-  input: model, scaler,  (pclass, sex, age, sibsp, parch, fare, embarked, initial)
-  output:  , 

```python
# predict_survival  
def predict_survival(model, scaler, pclass, sex, age, sibsp, parch, fare, embarked, initial):
    #    .
    #   ( input ) input  . 
    input_data = pd.DataFrame({
        'Pclass': [pclass],
        'Sex': [0 if sex == 'male' else 1],  #  .
        'Age': [age // 10],  #  10  .
        'SibSp': [sibsp],
        'Parch': [parch],
        'Fare': [fare],  #    .
        'Embarked': [1 if embarked == 'S' else (2 if embarked == 'C' else 3)],  #   .
        'Initial': [0 if initial == 'Mr' else (1 if initial == 'Miss' else (2 if initial == 'Mrs' else (3 if initial == 'Master' else 4)))],  #  .
        'Relatives': [sibsp + parch],
    })

    # 'Fare'  qcut  bins  .
    fare_bins = pd.qcut(df_org['Fare'], 9, retbins=True)[1]  # pd.qcut    .
    input_data['Fare'] = pd.cut(input_data['Fare'], bins=fare_bins, labels=False, include_lowest=True)

    #   .
    input_data_scaled = scaler.transform(input_data)
    
    #  .
    prediction = model.predict(input_data_scaled)
    prediction_proba = model.predict_proba(input_data_scaled)

    #   .
    result = "Survived" if prediction == 1 else "Not Survived"
    probability = prediction_proba[0][int(prediction)]  #    .
    
    return result, probability
```
